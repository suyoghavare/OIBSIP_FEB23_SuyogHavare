```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.pairplot(database)
```

```python
sns.pairplot(database1)
```

```python
plt.bar(database['Region'], database[' Estimated Unemployment Rate (%)'])
plt.xticks(rotation=90)
plt.xlabel('States')
plt.ylabel('Unemployment Rate')
plt.title('Unemployment according to States')
plt.show()
```

```python
sns.histplot(x=' Estimated Unemployment Rate (%)', data=database, kde=True, hue='Area')
plt.title('Unemployment according to Area')
plt.xlabel('Unemployment Rate')
plt.show()
```

```python
sns.lineplot(y=' Estimated Unemployment Rate (%)', x=' Date', data=database)
plt.title('Unemployment according to Date')
plt.xlabel('Date')
plt.xticks(rotation=90)
plt.ylabel('Unemployment Rate')
plt.show()
```

```python
plt.scatter(database1['Region.1'], database1[' Estimated Unemployment Rate (%)'])
plt.title('Unemployment according to Region')
plt.xlabel('Region')
plt.ylabel('Unemployment Rate')
plt.show()
```

```python
plt.bar(database['Region'], database[' Estimated Labour Participation Rate (%)'])
plt.xticks(rotation=90)
plt.xlabel('States')
plt.ylabel('Labour Participation Rate')
plt.title('Labour Participation according to States')
plt.show()
```

```python
sns.histplot(x=' Estimated Labour Participation Rate (%)', data=database, kde=True, hue='Area')
plt.title('Labour Participation according to Area')
plt.xlabel('Labour Participation Rate')
plt.show()
```

```python
plt.scatter(database1['Region.1'], database1[' Estimated Labour Participation Rate (%)'])
plt.title('Labour Participation according to Region')
plt.xlabel('Region')
plt.ylabel('Labour Participation Rate')
plt.show()
```
